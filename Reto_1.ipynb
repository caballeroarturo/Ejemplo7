```r
#setwd("C:/Users/cabal/Downloads")
#getwd()
#dir()

#altura <- read.csv("BD_Altura_Alunos.csv", sep = ";")
altura <- read.csv("https://raw.githubusercontent.com/beduExpert/Programacion-R-MASIVO-2021/main/Sesion-03/Reto-01/BD_Altura_Alunos.csv",
                   sep = ";")
View(altura)

head(altura)

# el histograma basico
hist(altura$Altura)

# le agregamos los breaks: es decir los cortes en el eje X
hist(altura$Altura,
     breaks = 20)

# podemos agregarle el titulo y el nombre de los ejes
hist(altura$Altura,
     breaks = 20,
     main = "Histograma de altura",
     xlab = "Altura",
     ylab = "Frecuencia")

#Se le agrega color de relleno de las barras
hist(altura$Altura,
     breaks = 20,
     main = " Histograma de alturas",
     ylab = "Frecuencia",
     xlab = "Altura",
     col = "lightblue")


altura %>%
  ggplot(aes(x=Altura)) +
  geom_histogram(binwidth = 2,
                 fill     = "#9ddaeb",
                 col      = "#f56642") +
  theme_classic() +
  labs(title    = "Histograma de altura",
       subtitle = "Personas de México",
       x = "Altura en cm",
       y = "Cantidad de personas en ese rango de altura")
```
