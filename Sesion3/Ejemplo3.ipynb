## Desarrollo
Realizamos un scatter plot de las variables wt y mpg, debemos utilizar necesariamente geom_point()

[, 1]  mpg   Miles/(US) gallon                        
       [, 2]  cyl   Number of cylinders                      
       [, 3]  disp  Displacement (cu.in.)                    
       [, 4]  hp    Gross horsepower                         
       [, 5]  drat  Rear axle ratio                          
       [, 6]  wt    Weight (1000 lbs)                        
       [, 7]  qsec  1/4 mile time                            
       [, 8]  vs    Engine (0 = V-shaped, 1 = straight)      
       [, 9]  am    Transmission (0 = automatic, 1 = manual)
       [,10]  gear  Number of forward gears                  
       [,11]  carb  Number of carburetors   

```r
g <- mtcars %>%
  ggplot(aes(x = wt, y = mpg)) +
    geom_point()
```

```r
g
```

modelo lineal, cambia el parametro `se`, este hace referencia al intervalo de confianza

```r
g <- g + geom_smooth(method = "lm", se = FALSE)
# modelo lineal, cambia el parametro `se`, este hace referencia al intervalo de confianza
```

```r
g
```

```r
g <- g + labs(x = 'Peso (x 1000 lbs)',
              y = 'Millas por Galón',
              title = "Gráfico de dispersión",
              subtitle = "Peso en libras VS Millas por galón")
```

```r
g
```

Agregar colores

```r
g <- mtcars %>%
  ggplot(aes(x = wt, y = mpg, col = cyl)) +
    geom_point() +
    labs(x = 'Weight (x1000lbs)',
         y = 'Miles per Gallon',
         colour = 'Number of\n Cylinders') # \n es salto de línea
```

```r
g
```

Se puede agregar la opción `facet_wrap()` con la variable `cyl`, para ***divirir la gráfica*** por el número de cilindros, esto ayudará a obtener información adicional

```r
table(mtcars$cyl)
```

Dividir la gráfica por columnas

```r
g + facet_wrap("cyl")
```

Al igual que el facet_wrap se puede dividir por tipo de transmisión utilizando la opción `facet_grid (am = Transmission (0 = automatic, 1 = manual))`

```r
g + facet_grid(am ~ cyl) # divide por renglones y columnas
```

# Resumen

```r
g <- mtcars %>%
  mutate(cyl = factor(cyl)) %>%
  ggplot(aes(x = wt, y = mpg, col = cyl)) +
    geom_point() +
    labs(x = 'Peso (x 1000 lbs)',
         y = 'Millas por Galón',
         title = "Gráfico de dispersión",
         subtitle = "Peso en libras VS Millas por galón",
         colour = 'Cantidad de\n Cilindros') + # \n es salto de línea
    facet_grid(am ~ cyl) # divide por renglones y columnas
```

```r
g
```
